```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("hardley_0327.csv")
```

```python
t = df["time"]
x1 = df["x1"]
y1 = df["y1"]
z1 = df["z1"]

# Plot
plt.figure(figsize=(12, 8), dpi=300)
plt.plot(t, x1, label="$x_1$", linewidth=1.5, color="green")
plt.plot(t, y1, label="$y_1$", linewidth=1.5, color="blue")
plt.plot(t, z1, label="$z_1$", linewidth=1.5, color="red")

plt.xlabel("t")
plt.ylabel("Values")
plt.title("Time evolution of $x_1$, $y_1$, and $z_1$")
plt.legend()
plt.grid(True)
plt.show()
```

```python
t, y1, y2, y3 = df["time"],df["y1"], df["y2"], df["y3"]


plt.figure(figsize=(12, 5))
plt.plot(y3, y2, linewidth=1.5, color="red")
plt.title("L80 Attractor (y3 vs y2)")
plt.xlabel("y3")
plt.ylabel("y2")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 5))
plt.plot(y1, y3, linewidth=1.5, color="blue")
plt.title("L80 Attractor (y1 x y3)")
plt.xlabel("y1")
plt.ylabel("y3")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 5))
plt.plot(y1, y2, linewidth=1.5, color="green")
plt.title("L80 Attractor (y1 x y2)")
plt.xlabel("y1")
plt.ylabel("y2")
plt.grid(True)
plt.show()
```
